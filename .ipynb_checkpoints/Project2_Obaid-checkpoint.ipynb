```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv("audit_risk.csv")
df2 = pd.read_csv("trial.csv")

### combine the two files in one dataframe
df = pd.concat([df1, df2], axis=1)
df.columns = map(str.lower, df.columns)
print(df.shape)

## remove duplicate columns 
df = df.loc[:,~df.columns.duplicated()]
print("the new dimensions are:",df.shape)
```

###  REMOVE MULTICOLLINEARITY 

```python

f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
## remove Para A , score A, para B, score B , number , score B1 , money value , score_MV ,prob
## history ,loss
df =  df.drop(['para_a','score_a','para_b','score_b','numbers','score_b.1',
               'money_value','score_mv','prob','history','loss','total','marks','district','district_loss'],axis=1)
```

```python
df = df[(df['location_id'] !='LOHARU')]
df = df[(df['location_id'] !='NUH')]
df = df[(df['location_id'] !='SAFIDON')]
```

```python
df.columns
```

```python
import seaborn as sns
%matplotlib inline 
sns.pairplot(df[['detection_risk', 'audit_risk','money_marks','risk_a','risk']],hue='risk')
```

```python
result=pd.get_dummies(df['location_id'],columns='location_id',prefix='location_id')
df=pd.concat([df, result], axis=1)
df = df.drop('location_id',axis=1)
```

# define y & X for Regression and split the data

```python
y= df['audit_risk']
X = df.drop(['audit_risk','risk'],axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0)  ## org stands for the very original
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)  
```

### Regression with bagging 

```python
### RANDOM FOREST regressor Bagging
from sklearn.ensemble import RandomForestRegressor

rnd_reg = RandomForestRegressor(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=0,bootstrap=True,max_depth=8)
rnd_reg.fit(X_train, y_train)

print(" R square for the training score ",rnd_reg.score(X_train,y_train))  
print(" R square for the testing score",rnd_reg.score(X_test,y_test))
```

```python
#### Bagging with bag Regressor 
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor(random_state=0)
bag_reg = BaggingRegressor(dt_reg, n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1, random_state=0)

bag_reg.fit(X_train, y_train)
print(" R square for the training score ",bag_reg.score(X_train,y_train))  
print(" R square for the testing score",bag_reg.score(X_test,y_test))
```

### Regression with Pasting

```python
### RANDOM FOREST regressor Pasting , 
from sklearn.ensemble import RandomForestRegressor

rnd_reg_pasting = RandomForestRegressor(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=0,bootstrap=False,max_depth=8)
rnd_reg_pasting.fit(X_train, y_train)

print(" R square for the training score ",rnd_reg_pasting.score(X_train,y_train))  
print(" R square for the testing score",rnd_reg_pasting.score(X_test,y_test))
```

```python
#### Pasting with bag Regressor 
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

dt_reg_pasting = DecisionTreeRegressor(random_state=0)
bag_reg_pasting = BaggingRegressor(dt_reg, n_estimators=500, max_samples=100, bootstrap=False, n_jobs=-1, random_state=0)

bag_reg_pasting.fit(X_train, y_train)
print(" R square for the training score ",bag_reg_pasting.score(X_train,y_train))  
print(" R square for the testing score",bag_reg_pasting.score(X_test,y_test))
```

### Regression with Adaptive boosting

```python

### Adaptive boosting with decision tree
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression

ada_reg1 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=6),n_estimators=200, learning_rate=0.5, random_state=0)

ada_reg1.fit(X_train, y_train)
print(" R square for the training score ",ada_reg1.score(X_train,y_train))  
print(" R square for the testing score",ada_reg1.score(X_test,y_test))
```

```python
## adaptive boosting with simple linear regression with polynomial features 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(2)   ### we will do adapative boosting with polynomial features
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

ada_reg2 = AdaBoostRegressor(LinearRegression(),n_estimators=200, learning_rate=0.5, random_state=0,)

ada_reg2.fit(X_train_poly, y_train)
print(" R square for the training score ",ada_reg2.score(X_train_poly,y_train))  
print(" R square for the testing score",ada_reg2.score(X_test_poly,y_test))
```

### GRADIENT BOOSTING REGRESSOR

```python
from xgboost import XGBRegressor
xgb_reg = XGBRegressor(max_depth=6,learning_rate=0.1,n_estimators=500 )

xgb_reg.fit(X_train, y_train)
print(" R square for the training score ",xgb_reg.score(X_train,y_train))  
print(" R square for the testing score",xgb_reg.score(X_test,y_test))
```

#### Lets do gridsearch and cross validation on xgboost Reg with k-fold 5

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 'max_depth':[4,6,8,10],
             'n_estimators':[50,100,200,300],
             'min_child_weight':[4,6,8,10,],
             'gamma':[0.001,0.01,1,10,100]}
grid_search = GridSearchCV(XGBRegressor(),param_grid, cv=5,return_train_score = True)

grid_search.fit(X_train,y_train)
```

### Regression with Adaptive boosting with PCA reduced data

```python
### Adaptive boosting with decision tree
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression

ada_reg1 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=6),n_estimators=200, learning_rate=0.5, random_state=0)

ada_reg1.fit(X_train_reduced, y_train)
print(" R square for the training score ",ada_reg1.score(X_train_reduced,y_train))  
print(" R square for the testing score",ada_reg1.score(X_test_reduced,y_test))
```

```python
## adaptive boosting with simple linear regression with polynomial features 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(2)   ### we will do adapative boosting with polynomial features
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

X_train_reduced_poly = pca.fit_transform(X_train_poly)
X_test_reduced_poly = pca.transform(X_test_poly)

ada_reg2 = AdaBoostRegressor(LinearRegression(),n_estimators=200, learning_rate=0.5, random_state=0,)

ada_reg2.fit(X_train_reduced_poly, y_train)
print(" R square for the training score ",ada_reg2.score(X_train_reduced_poly,y_train))  
print(" R square for the testing score",ada_reg2.score(X_test_reduced_poly,y_test))
```

----CONCLUSION FOR REGRESSION-----

IN regression most regressors explored performed poorly in testing set and poorly in testing. Use of PCA improved performance as it took some noise out. 

Best performance came from Adaptive boosting with linear regresion but features were converted to polynomial features with degree 2. This drastically improved the performance. 

Without turning features to polynomial features XGBOOST regressor gave the best score. 

#  GET DATA READY FOR CLASSIFICATION 

##  We will use classification report as the measuring metric

```python
y= df['risk']
X = df.drop(['audit_risk','risk'],axis =1)

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0)  ## org stands for the very original
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)  


from sklearn.metrics import classification_report
```

## Voting Classifier  two hard voting and two soft voting

```python
from sklearn.ensemble import VotingClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier

### to keep things simple we are using default values 
lg = LogisticRegression()
lg.fit(X_train, y_train)

knn = KNeighborsClassifier(7)
knn.fit(X_train, y_train)

svm = SVC(C = 10, probability = True)
svm.fit(X_train, y_train)

dct = DecisionTreeClassifier(random_state=0)
dct.fit(X_train,y_train)

xgb = XGBClassifier()
xgb.fit(X_train,y_train)
```

```python
voting1 = VotingClassifier(estimators=[('lg', lg), ('knn', knn), ('svc', svm)], voting='hard')
voting1.fit(X_train, y_train)

target_names = ['Risk 0', 'Risk 1']
y_pred = voting1.predict(X_test)
classification_report(y_test, y_pred, target_names=target_names)
```

```python
voting2 = VotingClassifier(estimators=[('dct', dct), ('xgb', xgb), ('svc', svm)], voting='hard')
voting2.fit(X_train, y_train)

target_names = ['Risk 0', 'Risk 1']
y_pred = voting2.predict(X_test)
classification_report(y_test, y_pred, target_names=target_names)
```

```python
voting3 = VotingClassifier(estimators=[('lg', lg), ('knn', knn), ('svc', svm)], voting='soft')
voting3.fit(X_train, y_train)

target_names = ['Risk 0', 'Risk 1']
y_pred = voting3.predict(X_test)
classification_report(y_test, y_pred, target_names=target_names)
```

```python
voting4 = VotingClassifier(estimators=[('dct', dct), ('xgb', xgb), ('svc', svm)], voting='soft')
voting4.fit(X_train, y_train)

target_names = ['Risk 0', 'Risk 1']
y_pred = voting4.predict(X_test)
classification_report(y_test, y_pred, target_names=target_names)
```

## Classification models with bagging

```python
from sklearn.ensemble import RandomForestClassifier

rnd = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=0,bootstrap=True,max_depth=8)
rnd.fit(X_train, y_train)

y_pred = rnd.predict(X_test)
classification_report(y_test, y_pred)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=0)
bag = BaggingClassifier(dt, n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1, random_state=0)

bag.fit(X_train, y_train)

y_pred = bag.predict(X_test)
classification_report(y_test, y_pred)
```

### Classification models with pasting 

```python
rnd2 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=0,bootstrap=False,max_depth=8)
rnd2.fit(X_train, y_train)

y_pred = rnd2.predict(X_test)
classification_report(y_test, y_pred)
```

```python
dt = DecisionTreeClassifier(random_state=0)
bag2 = BaggingClassifier(dt, n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1, random_state=0)

bag2.fit(X_train, y_train)

y_pred = bag2.predict(X_test)
classification_report(y_test, y_pred)
```

### Classifications models with Adaptive boosting 

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(DecisionTreeClassifier(max_depth=6),n_estimators=200, learning_rate=0.5, random_state=0)

ada.fit(X_train, y_train)

y_pred = ada.predict(X_test)
classification_report(y_test, y_pred)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, ada.decision_function(X_test))

close_zero = np.argmin(np.abs(thresholds))

plt.plot(precision[close_zero],recall[close_zero],'o',markersize=10,
        label='threshold zero', fillstyle='none' , c='k',mew=2)

plt.plot(precision,recall, label='precision recall label')
plt.xlabel('Precision')
plt.ylabel('recall')
plt.legend(loc='best')
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(SVC(C = 1, probability = True,gamma='auto'),n_estimators=50, learning_rate=0.5, random_state=0)

ada.fit(X_train, y_train)

y_pred = ada.predict(X_test)
classification_report(y_test, y_pred)
```

### Classification with Gradient Boosting

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth=6,learning_rate=0.1,n_estimators=300 ,gamma=10, min_child_weight=6)

xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)
y_probas = xgb.predict_proba(X_test)
classification_report(y_test, y_pred)
```
