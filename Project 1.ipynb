# Importing Data -->Concatenation --> initial analysis

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv("audit_risk.csv")
df2 = pd.read_csv("trial.csv")

### combine the two files in one dataframe
df = pd.concat([df1, df2], axis=1)
df.columns = map(str.lower, df.columns)
print(df.shape)

## remove duplicate columns 
df = df.loc[:,~df.columns.duplicated()]
print("the new dimensions are:",df.shape)
```

# Removing Multicolenarity 

```python
### check for collinerarity between these columns as they can have high correlations

f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```

```python
## remove Para A , score A, para B, score B , number , score B1 , money value , score_MV ,prob
## history ,loss
df =  df.drop(['para_a','score_a','para_b','score_b','numbers','score_b.1',
               'money_value','score_mv','prob','history','loss','total','marks','district','district_loss'],axis=1)
```

```python
####### based on feedback removing the non numeric values in Location_id 
df = df[(df['location_id'] !='LOHARU')]
df = df[(df['location_id'] !='NUH')]
df = df[(df['location_id'] !='SAFIDON')]
```

```python
print(df.shape) ## we are now reduced to 17 columns 
## convert location id to hot vector thru function dummy variables 
result=pd.get_dummies(df['location_id'],columns='location_id',prefix='location_id')
df=pd.concat([df, result], axis=1)
df = df.drop('location_id',axis=1)
```

```python
df.info() # showed zero object data types
#df.columns
```

# Split Data , Scale data and get ready for Regression

```python
y= df['audit_risk']
X = df.drop(['audit_risk','risk'],axis =1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error


x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0)  ## org stands for the very original
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)  
```

#### LINEAR REGRESSION WITH SCALED DATA 

```python
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()
lreg.fit(X_train,y_train)
print(" Rsquare for the training score ",lreg.score(X_train,y_train))  
print(" R square for the testing score",lreg.score(X_test,y_test))

myMSE =   mean_absolute_error(y_test, lreg.predict(X_test))  
print(myMSE)
print(y_test.mean())

```

#### LINEAR REGRESSION WITH RIDGE 

```python
from sklearn.linear_model import Ridge

x_range =[0.01,0.1,1,10]  ## my alpha values

train_score_list = []
test_score_list = []

for alpha in x_range:
    ridge = Ridge(alpha)
    ridge.fit(X_train,y_train)  ## takes in transformed data
    train_score_list.append(ridge.score(X_train,y_train))
    test_score_list.append(ridge.score(X_test,y_test))
    
print(train_score_list)
print(test_score_list)

myMSE =   mean_absolute_error(y_test, ridge.predict(X_test))  
print(myMSE)
```

```python
plt.plot(x_range,train_score_list,c='g',label = 'Train Score')
plt.plot(x_range,test_score_list,c='r',label = 'Test Score')
plt.xscale('log')
```

#### LINEAR REGRESSION WITH LASSO 

```python
from sklearn.linear_model import Lasso

x_range =[0.01,0.1,1,10]  ## my alpha values

train_score_list = []
test_score_list = []
meanerror = []

for alpha in x_range:
    lasso = Lasso(alpha)
    lasso.fit(X_train,y_train)  ## takes in transformed data
    train_score_list.append(lasso.score(X_train,y_train))
    test_score_list.append(lasso.score(X_test,y_test))
    meanerror.append(mean_absolute_error(y_test, lasso.predict(X_test))  )

print(train_score_list)
print(test_score_list)
print(meanerror)
```

```python
plt.plot(x_range,train_score_list,c='g',label = 'Train Score')
plt.plot(x_range,test_score_list,c='r',label = 'Test Score')
plt.xscale('log')
```

####  KNN REGRESSION

```python
from sklearn.neighbors import KNeighborsRegressor
x_range = [1,5,10,20]
train_score_list = []  ## initialize array to hold train values
test_score_list = []  ## initilaize array to hold test values
meanerror = []
for k in x_range:
    knn = KNeighborsRegressor(k)
    knn.fit(X_train,y_train)
    train_score_list.append(knn.score(X_train,y_train))
    test_score_list.append(knn.score(X_test,y_test))
    meanerror.append(mean_absolute_error(y_test, knn.predict(X_test))  )
    
print(train_score_list)
print(test_score_list)
print(meanerror)
```

####  POLYNOMIAL FEATURES REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures

train_score_list = []
test_score_list = []
meanerror = []
for n in range (1,4):
    poly = PolynomialFeatures(n)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    lreg.fit(X_train_poly,y_train)
    train_score_list.append(lreg.score(X_train_poly,y_train))
    test_score_list.append(lreg.score(X_test_poly,y_test))
    meanerror.append(mean_absolute_error(y_test, lreg.predict(X_test_poly))  )

print(train_score_list)
print(test_score_list)
print(meanerror)
## at n=2 gives the best results with train and test. Kinda overfit the data at train but test results come out good at n=2
```

# Get data Ready for classification, scale it and split the data

```python
## For classification things will be different 
y= df['risk']
X = df.drop(['audit_risk','risk'],axis =1)

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0)  ## org stands for the very original
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)  
```

```python
from sklearn.model_selection import cross_val_score 

from sklearn.model_selection import KFold
kfold = KFold(n_splits=5)

#kfold_scores = cross_val_score(knn,X_test, y_test, cv=kfold)
#print("K fold scores : {}".format(kfold_scores))
```

####  KNN CLASSIFICATION from k=1 - 10 

```python
## KNN classification 

from sklearn.neighbors import KNeighborsClassifier

train_score_list = []
test_score_list = []

for k in range(1,10):
    knn = KNeighborsClassifier(k)
    knn.fit(X_train,y_train)
    train_score_list.append( round(knn.score(X_train,y_train),3))
    test_score_list.append(  round(knn.score(X_test,y_test),3))

print("train scores are:  ", train_score_list)
print("test scores are: ",test_score_list)
```

```python
## lets plot them 
x_axis = range(1,10)
%matplotlib inline
plt.plot(x_axis, train_score_list,c='g',label='Training score')
plt.plot(x_axis, test_score_list,c='r',label='Testing score')

## extra info 
plt.xlabel('k value')
plt.ylabel('modelling accuracy')
```

####  CLASSIFICATION WITH LOGISTIC REGRESSION WITH C  clist = [0.01,0.1,1,10,100]  penality = 'l1'

```python
## logistic regression 
from sklearn.linear_model import LogisticRegression

clist = [0.01,0.1,1,10,100]
train_score_list = []
test_score_list = []


for c in clist:
    log_l1 = LogisticRegression(penalty='l1', C=c )
    log_l1.fit(X_train,y_train)
    train_score_list.append( round(log_l1.score(X_train,y_train),3))
    test_score_list.append( round(log_l1.score(X_test,y_test),3))

print("train scores are:  ", train_score_list)
print("test scores are: ",test_score_list)
```

####  Classification with LOGISTIC REGRESSION WITH PENALITY ='L2' &  clist = [0.01,0.1,1,10,100]  

```python
from sklearn.linear_model import LogisticRegression

clist = [0.01,0.1,1,10,100]
train_score_list = []
test_score_list = []


for c in clist:
    log_l2 = LogisticRegression(penalty='l2', C=c, solver='lbfgs')
    log_l2.fit(X_train,y_train)
    train_score_list.append(  round(log_l2.score(X_train,y_train),3))
    test_score_list.append(   round(log_l2.score(X_test,y_test),3))

print("train scores are:  ", train_score_list)
print("test scores are: ",test_score_list)
```
