```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

from sklearn.datasets import load_breast_cancer
from pandas.plotting import scatter_matrix 

data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y=data.target
```

```python
attributes = data.feature_names[0:5]
attributes
```

```python
scatter_matrix(X[attributes],figsize=(15,15),c=y,alpha=0.8,marker='0')
```

```python
X.info()
```

```python
X.describe()
```

```python

plt.scatter(X['mean area'],y,c=y)
plt.xlabel('Mean Area')
```

```python
plt.scatter(X['mean symmetry'],y,c=y)  ## understand that color = 'yellow'
plt.xlabel('Mean Symmetry')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

train_score_array = []  ## initialize array to hold train values
test_score_array = []  ## initilaize array to hold test values

for k in range(1,20):
    knn = KNeighborsClassifier(k)
    knn.fit(X_train,y_train)
    train_score_array.append(knn.score(X_train,y_train))
    test_score_array.append(knn.score(X_test,y_test))
```

```python
## lets plot them 
x_axis = range(1,20)
%matplotlib inline
plt.plot(x_axis, train_score_array,c='g',label='Training score')
plt.plot(x_axis, test_score_array,c='r',label='Testing score')

## extra info 
plt.xlabel('k value')
plt.ylabel('modelling accuracy')
```

# K Neighbors Regressor using boston data set

```python

from sklearn.datasets import load_boston


data = load_boston()

X = pd.DataFrame(data.data, columns=data.feature_names)
y=data.target
```

```python
attributes = data.feature_names[0:5]
attributes
```

```python
scatter_matrix(X[attributes],figsize=(15,15),c=y,alpha=0.8,marker='0')  ## instead of y you can also write data.
                                                                         ##target or an actual color
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

x_train_org,x_test_org,y_train,y_test=train_test_split(X,y,random_state=0)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(x_train_org)  ### you can fit and transform together 
X_test = scaler.transform(x_test_org)    #### just transform and not fit
```

```python
from sklearn.neighbors import KNeighborsRegressor
train_score_array = []  ## initialize array to hold train values
test_score_array = []  ## initilaize array to hold test values

for k in range(1,10):
    knn = KNeighborsRegressor(k)
    knn.fit(X_train,y_train)
    train_score_array.append(knn.score(X_train,y_train))
    test_score_array.append(knn.score(X_test,y_test))
```

```python
## lets plot them 
x_axis = range(1,10)
%matplotlib inline
plt.plot(x_axis, train_score_array,c='g',label='Training score regressor')
plt.plot(x_axis, test_score_array,c='b',label='Testing score regressor')

## extra info 
plt.xlabel('k value for regressor')
plt.ylabel('MSE')
```

```python
# CRIME and AGE 
# With only one neighbor as sklearn cannot handle more than tha
X_b = X_train[:50,6].reshape(-1,1)
y_b = y_train[:50]
```

```python
myknn=  KNeighborsRegressor(1)
myknn.fit(X_b,y_b)
```

```python
X_new = np.linspace(X_b.min(),X_b.max(),50).reshape(50,1)
y_predict = myknn.predict(X_new)

plt.plot(X_new,y_predict,c='r')
plt.scatter(X_b,y_b)
```
